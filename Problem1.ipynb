# Problem (Toffoli decomposition)

Prove the Toffoli gate decomposition

```python
import pennylane as qml

qml.drawer.use_style('pennylane')

def circ():
    qml.Toffoli(wires = (2,1,0))
    return 0

fig, ax = qml.draw_mpl(circ)()
```

```python
# matrix of the Toffoli gate in the computational basis

qml.matrix(qml.Toffoli((2,1,0)), wire_order  = (0,1,2))
```

```python
ops = qml.Toffoli.compute_decomposition((0,1,2))

tape = qml.tape.QuantumTape(ops)

fig, ax = qml.drawer.tape_mpl(tape)
```

```python
import numpy as np


np.round(
        qml.matrix(tape, wire_order  = (2,1,0)), 10).real
```
