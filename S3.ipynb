# The Solovay-Kitaev theorem

# Set G

*   $G  \subseteq \mathcal{SU}(2)$ is a finite set, $G = \{g_i: g_i^{-1}\in G\}$
* $G_l = \{g^{\alpha_1}_1g^{\alpha_2}_2\ldots g^{\alpha_l}_l: g_i \in G \,\,\text{and} \,\, \alpha_i=\pm 1\}$ is a set of *words* of length $l$
*  Let $⟨G⟩$ be the set of words of finite length.

* $⟨G⟩$ is dense in $ \mathcal{SU}(2)$

## Dense
$ G \subseteq  \mathcal{SU}(2)$ is said to be *dense* in $ \mathcal{SU}(2)$ if

$$
\forall \,\mathcal{U} \subseteq \mathcal{SU}(2) \, \, \text{and}  \, \,\epsilon >0, \,\,\exists g \in G: ||g-\mathcal{U}||<\epsilon
$$

# $\epsilon$-net


Let $\mathcal{N}, \mathcal{S} \subseteq \mathcal{SU}(2)$ then $\mathcal{N}$ is said to be an $\epsilon$-net for $\mathcal{S}$, where $ϵ > 0$, if

$$
\forall \,s \in \mathcal{S}, \,\,\exists \,n \in \mathcal{N}: ||n-s||<\epsilon
$$

# $\mathcal{S}_{\epsilon}$
$$\mathcal{S}_{\epsilon} =\{\mathcal{U} \subseteq \mathcal{SU}(2): ||u-\hat{1}||<\epsilon, \forall u \in \mathcal{U}  \}$$

# The Shrinking Lemma

$\exists$ $\,\,\epsilon'$  and $s: $ $ \forall \,\,G \subseteq \mathcal{SU}(2)$ and $\epsilon \leq \epsilon'$ we have:

if $G_l$ is a $\epsilon^2$-net for $\mathcal{S}_{\epsilon}$ $⇒$ $G_{5l}$ is an $s\epsilon^3$-net for $\mathcal{S}_{\sqrt{s\epsilon^3}}$.

The lemma allows to construct better approximations
at the cost of longer words and covering a smaller neighborhood of the identity.

# Corollary

There exist constants $\epsilon'$ and $s$ such that for any $G \in \mathcal{SU}(2)$, $\, \epsilon _0 \leq \epsilon'$  and $k \in Z_+$ we have that if
$G_{l_0}$ is an $\epsilon^2_0$-net for $\mathcal{S}_{\epsilon_0}$ then $G_{l_k}$ is an $\epsilon_k$-net for $\mathcal{S}_{\epsilon_k}$, where $\epsilon_k=(s\epsilon_0)^{(3/2)^k}$ and $l_k=5^{k}l_0$.
